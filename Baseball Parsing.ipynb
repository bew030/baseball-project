```python
# is it 4 for everything? 
important_links = [x.get('href') for x in all_links[4:-4]]
final_links = [link]
final_links = final_links+[base_link+x for x in important_links]
```

```python
# need to implement if there's a connection issue 
from time import sleep
def number_cycles(links, n=3):
    final_df = pd.DataFrame()
    for i in range(n):
        print('working on tab '+str(i))
        sleep(2)
        test = pd.read_html(links[i],header = 0)[0]
        final_df = pd.concat([final_df, test],ignore_index = True)
    return final_df
```

```python
df = number_cycles(final_links,len(final_links))
```

```python
df
```

```python
def dot_remover(x):
    if not (x == x):
        return x 
    else: 
        return x[2:]
```

```python
def dot_cleaner(df):
    df['Acquired'] = df['Acquired'].apply(dot_remover)
    df['Relinquished'] = df['Relinquished'].apply(dot_remover)
    return df
```

```python
no_dots_df = dot_cleaner(df)
```
