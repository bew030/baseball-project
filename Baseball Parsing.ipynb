```python
import json
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_link = 'https://www.prosportstransactions.com/baseball/Search/'
link = 'https://www.prosportstransactions.com/baseball/Search/SearchResults.php?Player=&Team=&BeginDate=2018-08-07&EndDate=2019-08-06&InjuriesChkBx=yes&submit=Search'
```

```python
r = requests.get(link)
urlText = r.text
```

```python
soup = BeautifulSoup(urlText, 'html.parser')
```

```python
all_links  = soup.find_all('a')
```

```python
all_links
```

```python
# is it 4 for everything? 
important_links = [x.get('href') for x in all_links[4:-4]]
final_links = ['https://www.prosportstransactions.com/baseball/Search/SearchResults.php?Player=&Team=&BeginDate=2018-08-07&EndDate=2019-08-06&InjuriesChkBx=yes&submit=Search']
final_links = final_links+[base_link+x for x in important_links]
```

```python
# need to implement if there's a connection issue 
from time import sleep
def number_cycles(links, n=3):
    final_df = pd.DataFrame()
    for i in range(n):
        print('working on tab '+str(i))
        sleep(2)
        test = pd.read_html(links[i],header = 0)[0]
        final_df = pd.concat([final_df, test],ignore_index = True)
    return final_df
```

```python
df = number_cycles(final_links,5)
```

```python
df
```

```python
def dot_remover(x):
    if not (x == x):
        return x 
    else: 
        return x[2:]
```

```python
def dot_cleaner(df):
    df['Acquired'] = df['Acquired'].apply(dot_remover)
    df['Relinquished'] = df['Relinquished'].apply(dot_remover)
    return df
```

```python
no_dots_df = dot_cleaner(df)
```

```python
no_dots_df['Date'] = pd.to_datetime(no_dots_df['Date'])
```
